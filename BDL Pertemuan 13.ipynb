```python
#BAR
import json
import pandas as pd
import pymongo
from bson.json_util import dumps
import matplotlib.pyplot as plt
myclient = pymongo.MongoClient("mongodb://127.0.0.1:27017") # server mongoDB
mydb = myclient["titanic"] # nama database
mycol = mydb["pessengers"] # nama collection
query = mycol.find({},{ "Age": 1, "_id": 0, "Name":1 })
list_cur = list(query)
df = pd.DataFrame(list_cur)
df = df.fillna(method='pad').head(20)
# Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(df["Name"], df["Age"].astype(float))
plt.xticks(df["Name"], rotation=90)
plt.ylabel("Umur")
plt.title("Umur Penumpang Titanic\n", size=16)
plt.show()
```

```python
import json
import pandas as pd
import pymongo
from bson.json_util import dumps
import matplotlib.pyplot as plt
import seaborn as sns
myclient = pymongo.MongoClient("mongodb://127.0.0.1:27017") # server mongoDB
mydb = myclient["titanic"] # nama database
mycol = mydb["pessengers"] # nama collection
query = mycol.find({},{ 'Age': 1, '_id': 0 })
list_cur = list(query)
df = pd.DataFrame(list_cur)
df = df.fillna(method='pad').head(1000)
# Seaborn
f, ax1 = plt.subplots(1, figsize=(5, 5), sharex=False)
sns.histplot(
 x=df['Age'].astype(float),
 stat= "frequency", #“count”, “frequency”, “density”, “probability”
 ax=ax1
)
ax1.set_title('Frekuensi Umur Penumpang Titanic\n', size=16)
sns.despine(bottom=True) # Remove the top and right spines from plot(s)
```
